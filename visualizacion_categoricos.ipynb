# Datos categoricos a Visualizar 

```python
# Importamos las librerias
import seaborn as sns 
import matplotlib.pyplot as plt

# Configuramos seaborn
sns.set_theme(style="ticks", context="talk", palette="tab10")


# Cargamos el dataset 
consumos = sns.load_dataset("tips")

# Configuramos el dataset en español
consumos.columns = ["total", "propina", "genero", "fumador", "dia", "tipo", "comensales"]
```

```python
consumos
```

# Graficas de dispersion

## Grafica striplot

```python
#sns.catplot(x="dia", y="total", data=consumos, jitter=True, kind="strip", hue="comensales", col="fumador")

sns.catplot(x="dia", y="total", data=consumos, kind="swarm", hue="comensales", col="fumador")

plt.show()
```

# Grafica categoricas de distribucion

## Grafica boxplot

```python
sns.catplot(x="dia", y="total", data=consumos,
            kind="box", col="fumador")
```

## Grafica boxenplot

```python
sns.catplot(x="dia", y="total", data=consumos,
            kind="boxen", col="fumador")
```

## Grafica violin

```python
sns.catplot(x="dia", y="total", data=consumos,
            kind="violin")
sns.catplot(x="dia", y="total", data=consumos,
            kind="violin", hue="fumador")
sns.catplot(x="dia", y="total", data=consumos,
            kind="violin", hue="fumador", split=True)
plt.show()
```

## Graficas categoricas de estimacion

## Grafica pointplot

```python
sns.catplot(x="dia", y="total", data=consumos,
            kind="point", col="fumador")
plt.show()
```

## Grafica barplot

```python
sns.catplot(x="dia", y="total", data=consumos,
            kind="bar", hue="comensales")
plt.show()
```

## Grafica countplot

```python
sns.catplot(x="dia", data=consumos,
            kind="count", hue="comensales")
plt.show()
```
