# Datos a Visualizar con Seaborn

```python
# Importamos las librerias necesarias
import matplotlib.pyplot as plt
import seaborn as sns
# importamos el conjunto de datos
pinguinos = sns.load_dataset("penguins")
# Lo cambiamos a español
pinguinos.columns = ["especie","isla","longitud_pico","profundidad_pico","longitud_aleta","masa","sexo"]
pinguinos
```

# Primera grafica de Seaborn

```python
# Ajustamos el tema
sns.set_theme(style="darkgrid", context="talk")
relacional = sns.relplot(data=pinguinos, height=7, aspect=1.5, markers=["s",".","P","d"])
relacional.set(title="Analisis de pinguinos")
relacional.set_ylabels("Variables")
relacional.set_xlabels("Numero de Piguino")
plt.show()
```

# Diagramas de dispercion

```python
print(">>>>>>   SIMPLE   <<<<<<<")
sns.relplot(data=pinguinos, x="longitud_pico", y="longitud_aleta" )

print(">>>>>>   MATIZ Y TAMAÑO FIJO   <<<<<<<")
sns.relplot(data=pinguinos, x="longitud_pico", y="longitud_aleta", hue="masa", s=50)
sns.relplot(data=pinguinos, x="longitud_pico", y="longitud_aleta", hue="especie", s=50)

print(">>>>>>   REJILLA, TAMAÑO PERSONALIZADO Y ESTILO   <<<<<<<")
sns.relplot(data=pinguinos, x="longitud_pico", y="longitud_aleta", hue="especie", row="sexo", size="masa", sizes=(100,500), style="sexo")
sns.relplot(data=pinguinos, x="longitud_pico", y="longitud_aleta", hue="especie", row="sexo", col="isla", size="masa", sizes=(100,500), style="sexo")


```

# Grafico de Lineas

```python
sns.relplot(data=pinguinos, x="especie", y="masa", kind="line", hue="sexo", style="sexo")
plt.show()

sns.relplot(data=pinguinos, x="profundidad_pico", y="longitud_aleta", kind="line", hue="especie", style="especie")
plt.show()
```

# Combinacion de graficos de lineas y de dispercion

```python
figura,ejes = plt.subplots(1, 2, figsize=(15, 5))
sns.lineplot(data=pinguinos,
            x="especie",
            y="masa",
            hue="sexo",
            style="sexo",
            ax=ejes[0])
sns.scatterplot(data=pinguinos,
            x="longitud_pico",
            y="longitud_aleta",
            hue="especie",
            s=300,
            ax=ejes[1])
plt.show()
```
